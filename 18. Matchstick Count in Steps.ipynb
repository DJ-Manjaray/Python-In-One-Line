## MATCHSTICK COUNT IN STEPS

Count of matchsticks required to represent the given number

```python
def match_houses(steps):
    return steps * 6 - (steps - 1) if steps > 0 else 0
```

```python
match_houses(1)
```

```python
match_houses(6)
```
